```python
import python.reuse_model as model
import python.utility as util
```

```python
# hd1 = [complex(1,-1)]
# hd2 = [complex(0,1)]

hd1 = [complex(-1,-1)]
hd2 = [complex(1,1)]

# hd1 = [1,complex(1,1),complex(0,1), complex(1,-1)]
# hd2 = [1,complex(-1,-1),2, complex(0,1)]

snn = model.FHRR_SNN_Model(len(hd1))
```

```python
snn_rslt, voltages = snn.bind(hd1, hd2, True)
```

```python
py_rslt = snn.bind_py(hd1, hd2)
print(util.convert_rad2deg(util.complexarr2phase(snn_rslt)))
print(util.convert_rad2deg(util.complexarr2phase(py_rslt)))
```

```python
print(snn.similarity(py_rslt, snn_rslt))
```

## Reuse

```python
snn_rslt, voltages = snn.bind(hd2, hd1, True)
py_rslt = snn.bind_py(hd2, hd1)
print(util.convert_rad2deg(util.complexarr2phase(snn_rslt)))
print(util.convert_rad2deg(util.complexarr2phase(py_rslt)))
print(snn.similarity(py_rslt, snn_rslt))
py_rslt = snn.bind_py(hd2, hd1)
print(util.convert_rad2deg(util.complexarr2phase(snn_rslt)))
print(util.convert_rad2deg(util.complexarr2phase(py_rslt)))
print(snn.similarity(py_rslt, snn_rslt))
```

```python
titles = ["voltage_n1", "voltage_n2", "voltage_mult", "voltage_add", "voltage_deco_mult", "voltage_deco_add"]

util.show_voltages(titles, voltages, figsize=(15,20))
```

## Reuse for bundling

```python
snn_rslt, voltages = snn.bundle(hd1, hd2, True)
py_rslt = snn.bundle_py(hd1, hd2)
print(util.convert_rad2deg(util.complexarr2phase(snn_rslt)))
print(util.convert_rad2deg(util.complexarr2phase(py_rslt)))
print(snn.similarity(py_rslt, snn_rslt))
```

```python
titles = ["voltage_n1", "voltage_n2", "voltage_mult", "voltage_add", "voltage_deco_mult", "voltage_deco_add"]

util.show_voltages(titles, voltages, figsize=(15,20))
```
