```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst_range = 100
thresh    = 20
x = [i for i in range(lst_range) ]
y = []
voltage = 0
for i in range(lst_range):
    if x[i] % thresh == 0:
        voltage = 0
    else:
        voltage += 2
    y.append(voltage)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
'''
Find gradient of a [x], [y] data paired 2 lists
'''
import random
def find_gradient(x, y):
    if (len(x) != len(y)):
        assert("Dimensions does not match")
    lst_size = len(x)
    gradients = {}
    for i in range(1, lst_size-1):
        m1 = 0
        m2 = 0
        rnd_int = i
        x1 = x[rnd_int]
        x2 = x[rnd_int + 1]
        x3 = x[rnd_int - 1]
        m1 = (y[rnd_int + 1] - y[rnd_int])     / (x2 - x1)
        m2 = (y[rnd_int    ] - y[rnd_int - 1]) / (x1 - x3)
        if (round(m1,2) == round(m2,2)):
            if (m1 > 0):
                gradients[round(m1,2)] = i
    return list(gradients.keys())
```
